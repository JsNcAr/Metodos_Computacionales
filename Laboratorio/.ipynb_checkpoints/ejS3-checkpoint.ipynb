```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h=0.01
t = np.arange(0,10,h)
rx = (0.28*t)+(0.036*(t**2))


forward_x= np.diff(rx)/h

exact_x = 0.28+(2*0.036*t)



plt.plot(t[:-1:], forward_x)
plt.plot(t[:-1:], exact_x[:-1:])
plt.show()
```

```python
ry = 0.019*(t**3)

forward_y = np.diff(ry)/h

exact_y = 0.059*(t**2)

plt.plot(t[:-1:], forward_y)
plt.plot(t[:-1:], exact_y[:-1:])
plt.show()
```
