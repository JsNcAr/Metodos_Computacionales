```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sec_dev(f,x,h):
    return (f(x+h)+f(x-h)-2*f(x))/(h**2)

def cos(x):
    return np.cos(x)
def dev_teo(x):
    return (-np.cos(x))
```

```python
h=0.01
x = np.arange(0,np.pi,h)

x_e = sec_dev(cos,x,h)
x_t = dev_teo(x)
```

```python
plt.plot(x, x_e, "o")
plt.plot(x, x_t)
plt.show()
```

```python
exp = sec_dev(cos,np.pi/4,h)
teo = dev_teo(np.pi/4)
print(exp, teo, exp-teo)
```
