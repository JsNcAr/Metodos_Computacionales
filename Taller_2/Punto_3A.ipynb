a) Carga encerrada =  $ 2\pi\alpha\int_{r1}^{r2}r^2dr$



```python
r1 = 0.10
r2 = 0.20
alpha = 0.5

N1 = 81

h1 = (r2-r1)/(N1-1)

array = np.linspace(r1, r2, N1)

r = pow(array,2) 

Deltar = round(0.5*h1*(sum(r[0:N1-1]) + sum(r[1:N1-1])),4)

Qtotal = 2*np.pi*alpha*Deltar
```
